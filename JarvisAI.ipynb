```python
import pyttsx3
import speech_recognition as sr
import wikipedia
import webbrowser
from datetime import datetime
from googletrans import Translator
import tkinter as tk
from tkinter import scrolledtext
import threading
from youtube_search import YoutubeSearch
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# Initialize the translator
translator = Translator()

# Initialize the text-to-speech engine
engine = pyttsx3.init()

# Initialize the speech recognizer
r = sr.Recognizer()

# Initialize the web driver (make sure you have Chrome and chromedriver installed)
driver = webdriver.Chrome()

def say(text, lang='en'):
    voices = engine.getProperty('voices')
    
    if lang == 'en':
        engine.setProperty('voice', voices[1].id)  # English voice
    else:
        for voice in voices:
            if lang in voice.languages:
                engine.setProperty('voice', voice.id)
                break
    
    engine.say(text)
    engine.runAndWait()

def takecommand():
    with sr.Microphone() as source:
        r.pause_threshold = 1
        print("Listening...")
        update_ui("Listening...")
        audio = r.listen(source)
        try:
            query = r.recognize_google(audio, language="en-in")
            print(f"User said: {query}")
            update_ui(f"User said: {query}")
            return query
        except sr.UnknownValueError:
            print("Sorry, I didn't catch that.")
            update_ui("Sorry, I didn't catch that.")
            return None
        except sr.RequestError:
            print("Sorry, there was an error with the speech recognition service.")
            update_ui("Sorry, there was an error with the speech recognition service.")
            return None

def get_wikipedia_summary(topic):
    summary = wikipedia.summary(topic, sentences=2)
    return summary

def translate_text(text, dest_lang='en'):
    translation = translator.translate(text, dest=dest_lang)
    return translation.text

def update_ui(message):
    text_area.insert(tk.END, message + "\n")
    text_area.see(tk.END)

def play_youtube_video(query):
    try:
        results = YoutubeSearch(query, max_results=1).to_dict()
        if results:
            video_url = f"https://www.youtube.com/watch?v={results[0]['id']}"
            driver.get(video_url)
            
            # Wait for the video player to load
            wait = WebDriverWait(driver, 10)
            play_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button.ytp-play-button")))
            play_button.click()
            
            say(f"Now playing {results[0]['title']}")
            update_ui(f"Now playing {results[0]['title']}")
        else:
            say("Sorry, I couldn't find a video matching your request.")
    except Exception as e:
        print(f"An error occurred: {e}")
        update_ui(f"An error occurred: {e}")
        say("Sorry, I couldn't play that video.")

def run_assistant():
    say("Hello, I am Jarvis")
    
    while True:
        user_query = takecommand()
        
        if user_query:
            if 'time' in user_query:
                current_time = datetime.now().strftime('%H:%M')
                say(f"The current time is {current_time}")
            elif 'date' in user_query:
                current_date = datetime.now().strftime('%Y-%m-%d')
                say(f"Today's date is {current_date}")
            elif 'Wikipedia' in user_query:
                topic = user_query.replace('Wikipedia', '').strip()
                summary = get_wikipedia_summary(topic)
                say(summary)
            elif 'open' in user_query and 'play' in user_query:
                search_query = user_query.replace('open', '').replace('play', '').strip()
                play_youtube_video(search_query)
            elif 'translate' in user_query:
                say("What would you like me to translate?")
                text_to_translate = takecommand()
                if text_to_translate:
                    say("To which language should I translate?")
                    dest_lang = takecommand()
                    if dest_lang:
                        try:
                            translated_text = translate_text(text_to_translate, dest_lang)
                            say(f"The translation is: {translated_text}")
                        except ValueError:
                            say("Sorry, I couldn't recognize that language. Please try again.")
            elif 'stop' in user_query or 'exit' in user_query or 'quit' in user_query:
                say("Goodbye!")
                driver.quit()
                root.quit()
                break
            else:
                say("I didn't understand that command.")

# Create the main window
root = tk.Tk()
root.title("Jarvis Assistant")
root.geometry("400x400")

# Create a scrolled text area
text_area = scrolledtext.ScrolledText(root, wrap=tk.WORD, width=40, height=20)
text_area.pack(padx=10, pady=10)

# Create a button to start the assistant
start_button = tk.Button(root, text="Start Jarvis", command=lambda: threading.Thread(target=run_assistant).start())
start_button.pack(pady=10)

# Start the UI
root.mainloop()
```

```python
import pyttsx3
import speech_recognition as sr
import wikipedia
import webbrowser
from datetime import datetime
from googletrans import Translator
import tkinter as tk
from tkinter import scrolledtext
import threading
from youtube_search import YoutubeSearch
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import openai
import os

# Initialize the OpenAI API
openai.api_key = os.getenv("OPENAI_API_KEY")

# Initialize the translator
translator = Translator()

# Initialize the text-to-speech engine
engine = pyttsx3.init()

# Initialize the speech recognizer
r = sr.Recognizer()

# Initialize the web driver (make sure you have Chrome and chromedriver installed)
driver = webdriver.Chrome()

# Language codes dictionary
LANGUAGE_CODES = {
    'german': 'de',
    'french': 'fr',
    'spanish': 'es',
    'italian': 'it',
    'japanese': 'ja',
    'chinese': 'zh-cn',
    'russian': 'ru',
    # Add more languages as needed
}

def say(text, lang='en'):
    voices = engine.getProperty('voices')
    
    if lang == 'en':
        engine.setProperty('voice', voices[1].id)  # English voice
    else:
        for voice in voices:
            if lang in voice.languages:
                engine.setProperty('voice', voice.id)
                break
    
    engine.say(text)
    engine.runAndWait()

def takecommand():
    with sr.Microphone() as source:
        r.pause_threshold = 1
        print("Listening...")
        update_ui("Listening...")
        audio = r.listen(source)
        try:
            query = r.recognize_google(audio, language="en-in")
            print(f"User said: {query}")
            update_ui(f"User said: {query}")
            return query
        except sr.UnknownValueError:
            print("Sorry, I didn't catch that.")
            update_ui("Sorry, I didn't catch that.")
            return None
        except sr.RequestError:
            print("Sorry, there was an error with the speech recognition service.")
            update_ui("Sorry, there was an error with the speech recognition service.")
            return None

def get_wikipedia_summary(topic):
    summary = wikipedia.summary(topic, sentences=2)
    return summary

def translate_text(text, dest_lang='en'):
    try:
        # Convert dest_lang to lowercase and get the corresponding code
        dest_lang = dest_lang.lower()
        lang_code = LANGUAGE_CODES.get(dest_lang, dest_lang)
        
        # Detect the source language
        src_lang = translator.detect(text).lang
        
        # Translate the text
        translation = translator.translate(text, src=src_lang, dest=lang_code)
        
        return translation.text
    except Exception as e:
        print(f"Translation error: {e}")
        return None

def update_ui(message):
    text_area.insert(tk.END, message + "\n")
    text_area.see(tk.END)

def play_youtube_video(query):
    try:
        results = YoutubeSearch(query, max_results=1).to_dict()
        if results:
            video_url = f"https://www.youtube.com/watch?v={results[0]['id']}"
            driver.get(video_url)
            
            # Wait for the video player to load
            wait = WebDriverWait(driver, 10)
            play_button = wait.until(EC.element_to_be_clickable((By.CSS_SELECTOR, "button.ytp-play-button")))
            play_button.click()
            
            say(f"Now playing {results[0]['title']}")
            update_ui(f"Now playing {results[0]['title']}")
        else:
            say("Sorry, I couldn't find a video matching your request.")
    except Exception as e:
        print(f"An error occurred: {e}")
        update_ui(f"An error occurred: {e}")
        say("Sorry, I couldn't play that video.")

def get_openai_response(prompt):
    try:
        response = openai.Completion.create(
            engine="text-davinci-002",
            prompt=prompt,
            max_tokens=150
        )
        return response.choices[0].text.strip()
    except Exception as e:
        print(f"An error occurred with OpenAI API: {e}")
        return None

def run_assistant():
    say("Hello, I am Jarvis")
    
    while True:
        user_query = takecommand()
        
        if user_query:
            if 'time' in user_query:
                current_time = datetime.now().strftime('%H:%M')
                say(f"The current time is {current_time}")
            elif 'date' in user_query:
                current_date = datetime.now().strftime('%Y-%m-%d')
                say(f"Today's date is {current_date}")
            elif 'Wikipedia' in user_query:
                topic = user_query.replace('Wikipedia', '').strip()
                summary = get_wikipedia_summary(topic)
                say(summary)
            elif 'open' in user_query and 'play' in user_query:
                search_query = user_query.replace('open', '').replace('play', '').strip()
                play_youtube_video(search_query)
            elif 'translate' in user_query.lower():
                say("What would you like me to translate?")
                text_to_translate = takecommand()
                if text_to_translate:
                    say("To which language should I translate?")
                    dest_lang = takecommand()
                    if dest_lang:
                        translated_text = translate_text(text_to_translate, dest_lang)
                        if translated_text:
                            say(f"The translation is: {translated_text}")
                            update_ui(f"Translated to {dest_lang}: {translated_text}")
                        else:
                            say(f"Sorry, I couldn't translate to {dest_lang}. Please try again with a different language.")
                            update_ui(f"Translation to {dest_lang} failed.")
            elif 'stop' in user_query or 'exit' in user_query or 'quit' in user_query:
                say("Goodbye!")
                driver.quit()
                root.quit()
                break
            else:
                openai_response = get_openai_response(user_query)
                if openai_response:
                    say(openai_response)
                else:
                    say("I'm sorry, I couldn't process that request.")

# Create the main window
root = tk.Tk()
root.title("Jarvis Assistant")
root.geometry("400x400")

# Create a scrolled text area
text_area = scrolledtext.ScrolledText(root, wrap=tk.WORD, width=40, height=20)
text_area.pack(padx=10, pady=10)

# Create a button to start the assistant
start_button = tk.Button(root, text="Start Jarvis", command=lambda: threading.Thread(target=run_assistant).start())
start_button.pack(pady=10)

# Start the UI
root.mainloop()
```
